# 1. Quais são os imoveis que a House Rockets deveria comprar e por qual preço?
### Plano 1:
    - Coletar dados do Kaggle
    - Agrupar os dados por região (Zipcode)
    - Dentro de cada região, estudar estatisticamente os preços dos imóveis
    - Responder a pergunta com base em um valor encontrar em comum (Ex: mediana)
    - Condições de compra:
        1. Se condition maior que 3 e preco de compra menor que mediana da regiao
        Comprar
        
        2. Se condition igual a 2 e preco de compra menor que mediana da regiao - 20%
        Comprar 
        
        3. Se condition menor que 2 e preco de compra menor que mediana da regiao - 40%
        Comprar 
        
        4. Outros casos, não comprar
## Exemplo:
Codigo_imovel | Regiao | Preco_do_imovel | Preco_mediana | condicao | Status

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:/Users/lucas/Estudos/_CDS/repos/python_zero_ds/data/kc_house_data.csv')
df
```

```python
df.columns
```

```python
df_median_region = df[['zipcode', 'price']].groupby('zipcode').median().reset_index()
df_median_region.rename(columns={'price':'median_price_by_region'}, inplace=True)
```

```python
df_final1 = pd.merge(df, df_median_region, how='inner', on='zipcode')
df_final1 = df_final1[['id','zipcode', 'price', 'median_price_by_region', 'condition']]
df_final1['status'] = df_final1[['price', 'median_price_by_region', 'condition']].apply(lambda x: 
                                                                                      'Comprar' if x['condition'] >= 3 and x['price'] < x['median_price_by_region'] else 
                                                                                      'Comprar' if x['condition'] == 2 and x['price'] < x['median_price_by_region']*(1-0.2) else
                                                                                      'Comprar' if x['condition'] < 2 and x['price'] < x['median_price_by_region']*(1-0.4) else 'Nao comprar', axis=1)
df_final1
```

# 2. Uma vez o imóvel comprado, qual o melhor momento para vende-lo, e por qual valor
## Plano 1:
    - Adicionar coluna de sazonalidade(season) com base na data(date)
    - Agrupar os dados por região (zipcode) e por sazonalidade(season)
    - Dentro de cada região e sazonalidade, calcular a mediana
    - Condições de venda:
        1. Se o preço de compra for maior que a mediana da região + sazonalidade
        Preço da venda será igual ao preço da compra + 10%
        
        2. Se o preço de compra for menor que a mediana da região + sazonalidade
        Preço da venda será igual ao preço da compra + 30%
        
        
## Exemplo:
Codigo_imovel | Regiao | Temporada(sazonalidade) | Preco_mediana | Preco_compra | preco_venda | lucro

```python
df_aux = df
```

```python
df_aux['season'] = df_aux['date'].apply(lambda x: 'Winter' if int(x[4:6]) in [12, 1, 2] else
                                            'Spring' if int(x[4:6]) in [3, 4, 5] else
                                            'Summer' if int(x[4:6]) in [6, 7, 8] else 
                                            'Autumn' if int(x[4:6]) in [9, 10, 11] else 'Out of date')
```

```python
df_region_season = df_aux[['zipcode', 'season', 'price']].groupby(['zipcode', 'season']).median().reset_index()
df_region_season.rename(columns={'price': 'median_price_by_region_season'}, inplace=True)
```

```python
df_final2 = pd.merge(df, df_region_season, how='inner', on='zipcode')
```

```python
df_final2['status'] = df_region_season[['price', 'median_price_by_region_season',]].apply(lambda x: 
                                                                                        'Comprar' if x['price'] < x['median_price_by_region_season'] else
                                                                                        'Comprar' if x['price'] < x['median_price_by_region_season'] else)
```
