## 1. Crie uma coluna chamada 'house_age'
    - Se a data da coluna 'date' for maior ou igual que 2014-01-01 : 'new_house'
    - Se a data da coluna 'date' for menor que 2014-01-01 : 'old_house'

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df['house_age'] = df['date'].apply(lambda x: 'new_house' if x >= pd.to_datetime('2014-01-01') else 'old_house')
df.head()
```

## 2. Crie uma coluna chamada 'dormitory_type'
    - Se o valor da coluna 'bedrooms' for == 1 : 'studio'
    - Se o valor da coluna 'bedrooms' for == 2 : 'apartment'
    - Se o valor da coluna 'bedrooms' for > 2 : 'house'

```python
# Verificando se nao existem valores negativos ou 0
df['bedrooms'].unique()
```

```python
df['dormitory_type'] = df['bedrooms'].apply(lambda x: 'studio' if x == 1 else 
                                                      'apartment' if x == 2 else 
                                                      'house' if x > 2 else 'NA') # por ter valores = 0
df.head()
```

```python
df[df['dormitory_type'] == 'NA']
```

## 3. Crie uma coluna chamada 'condition_type'
    - Se o valor da coluna 'condition' for <= 2 : 'bad'
    - Se o valor da coluna 'condition' for == [3,4] : 'regular'
    - Se o valor da coluna 'condition' for == 5 : 'good'

```python
df['condition'].unique()
```

```python
df['condition_type'] = df['condition'].apply(lambda x: 'bad' if x <= 2 else
                                                       'regular' if x == 3 or x == 4 else 'good' )
df.head()
```
