```python
import ipywidgets as widgets
from ipywidgets import fixed, Layout
```

```python
style = {'description_width': 'initial'}

living_sqft_limit = widgets.IntSlider(
    value = int(df['sqft_living'].mean()), # valor fixo
    min = int(df['sqft_living'].min()),
    max = int(df['sqft_living'].max()),
    step = 10, # quanto varia
    description = 'Minimum sqft living',
    disable = False, # para aparecer
    style = style)

bathroom_limit = widgets.FloatSlider(
    value = float(df['bathrooms'].mean()), # valor fixo
    min = float(df['bathrooms'].min()),
    max = float(df['bathrooms'].max()),
    step = 1, # quanto varia
    description = 'Minimum bathrooms',
    disable = False, # para aparecer
    style = style)

max_price = widgets.IntSlider(
    value = int(df['price'].mean()), # valor fixo
    min = int(df['price'].min()),
    max = int(df['price'].max()),
    step = 1000, # quanto varia
    description = 'Maximum price',
    disable = False, # para aparecer
    style = style)


condition_limit = widgets.IntSlider(
    value = 2, # valor fixo
    min = df['condition'].min(),
    max = df['condition'].max(),
    step = 1, # quanto varia
    description = 'Condition limit',
    disable = False, # para aparecer
    style = style)

yr_built_limit = widgets.IntSlider(
    value = df['yr_built'].mean(), # valor fixo
    min = df['yr_built'].min(),
    max = df['yr_built'].max(),
    step = 1, # quanto varia
    description = 'Year built limit',
    disable = False, # para aparecer
    style = style)
```

```python
import plotly.express as px
```

```python
def update_map(df, max_price, bathroom_limit, living_sqft_limit, condition_limit, yr_built_limit):
    houses = df[(df['price'] <= max_price) & 
                (df['bathrooms'] >= bathroom_limit) & 
                (df['sqft_living'] >= living_sqft_limit) & 
                (df['condition'] >= condition_limit) & 
                (df['yr_built'] >= yr_built_limit)][['id', 'lat', 'long', 'price', 'bathrooms', 'sqft_living', 'condition', 'yr_built']]
    
    fig = px.scatter_mapbox(houses,
                       lat='lat',
                       lon='long',
                       hover_data=['bathrooms', 'sqft_living', 'condition', 'yr_built'],
                       color_continuous_scale=px.colors.cyclical.IceFire,
                       size='price',
                       zoom=10)

    fig.update_layout(mapbox_style='open-street-map')
    fig.update_layout(height=600, margin={'r':0, 't':0, 'l':0, 'b':0})
    fig.show()
```

```python
widgets.interactive(update_map, df=fixed(df), max_price = max_price, 
                                              bathroom_limit = bathroom_limit, 
                                              living_sqft_limit = living_sqft_limit, 
                                              condition_limit = condition_limit, 
                                              yr_built_limit = yr_built_limit)
```

## 5. adicione os seguintes filtros no dashboard
- filtro por data disponivel de compra
- filtro por ano de renovacao
- filtro se possui vista para a agua ou nao

```python
# formatação dados
df['year'] = pd.to_datetime(df['date']).dt.strftime('%Y')
df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')
df['year_week'] = pd.to_datetime(df['date']).dt.strftime('%Y-%U')
```

```python
date_limit = widgets.SelectionSlider(
    options=df['date'].sort_values().unique().tolist(),
    value='2014-12-01',
    description='Max available date',
    disable=False,
    continuous_update=False,
    style=style,
    readout=True)

year_limit = widgets.SelectionSlider(
    options=df['yr_renovated'].sort_values().unique().tolist(),
    value=2000,
    description='Max Year',
    disable=False,
    continuous_update=False,
    style=style,
    readout=True)

waterfront_limit = widgets.Checkbox(
    value=False,
    description='Is Waterfront?',
    disable=False,
    indent=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns
```

```python
def update_map(df, date_limit, year_limit, waterfront_limit):
    houses = df[(df['date'] < date_limit) & 
                (df['yr_renovated'] >= year_limit) & 
                (df['waterfront'] == waterfront_limit)].copy()
    
    fig = plt.figure(figsize=(24, 12))
    specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
    
    ax1 = fig.add_subplot(specs[0, :])
    ax2 = fig.add_subplot(specs[1, 0])
    ax3 = fig.add_subplot(specs[1, 1])

    by_year = houses[['id', 'year']].groupby('year').sum().reset_index()
    sns.barplot(x='year', y='id', data=by_year, ax=ax1)
    
    by_day = houses[['id', 'date']].groupby('date').sum().reset_index()
    sns.lineplot(x='date', y='id', data=by_day, ax=ax2)
    plt.xticks(rotation=60)
    
    by_week_of_year = houses[['id', 'year_week']].groupby('year_week').mean().reset_index()
    sns.barplot(x='year_week', y='id', data=by_week_of_year, ax=ax3)
    plt.xticks(rotation=60)
```

```python
widgets.interactive(update_map, df=fixed(df), date_limit = date_limit, 
                                              year_limit = year_limit, 
                                              waterfront_limit = waterfront_limit)
```
